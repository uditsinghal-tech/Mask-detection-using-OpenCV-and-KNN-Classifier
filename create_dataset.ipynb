```python
import cv2
```

```python
img=cv2.imread("image.jpg")
```

```python
img.shape
```

```python
img
```

```python
haar_data=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
haar_data.detectMultiScale(img)
```

```python
#Drwing rectangle on all the detected faces
while True:
    faces=haar_data.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
    cv2.imshow("Face",img)
    if cv2.waitKey(2) == 27:
        break
cv2.destroyAllWindows()
```

```python
#By the help of webcam a dataset of images without mask is created consisting of 200 images
capture=cv2.VideoCapture(0)
data_without_mask=[]
while True:
    flag,img=capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
            face=img[y:y+h, x:x+w, :]
            face=cv2.resize(face,(50,50))
            print(len(data_without_mask))
            if len(data_without_mask)<200:
                data_without_mask.append(face)
        cv2.imshow("Face without mask",img)
        if cv2.waitKey(2) == 27 or len(data_without_mask)>=200:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
```

```python
np.save('without_the_mask.npy',data_without_mask)
```

```python
#By the help of dataset with mask is created consisting of 200 images
capture=cv2.VideoCapture(0)
data_with_mask=[]
while True:
    flag,img=capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
            face=img[y:y+h, x:x+w, :]
            face=cv2.resize(face,(50,50))
            print(len(data_with_mask))
            if len(data_with_mask)<200:
                data_with_mask.append(face)
        cv2.imshow("Face with mask",img)
        if cv2.waitKey(2) == 27 or len(data_with_mask)>=200:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
np.save('with_the_mask.npy',data_with_mask)
```
